```python
# Importing all the required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
# Loading the Dataset
df = pd.read_csv("british_airways_reviews_extended.csv")

# head function prints the top 5 rows of the Dataset.
df.head()
```

```python
# Checking for the NULL values count in the Dataset.
df.isnull().count()
```

```python
# Checking all the Columns and Datatypes
df.info()
```

```python
# Renamking the Column from date to review_date
df = df.rename(columns={'date':'review_date'})

df.head()
```

```python
# Converting all the Column names to lower
df.columns = df.columns.str.lower()

df.head()
```

```python
# Dropping the Unnecessary Columns
df.drop(columns=['trip_verified', 'helpful_votes'], inplace=True)

df.head()
```

```python
# Remove "✅ Trip Verified" and "Not Verified" from reviews
df["reviews"] = df["reviews"].str.replace(r"✅\s*Trip Verified\s*\|?\s*", "", regex=True)
df["reviews"] = df["reviews"].str.replace(r"Not Verified\s*\|?\s*", "", regex=True)

# Show the first few cleaned reviews
df.head(10)
```

```python
# Function to clean ordinal suffixes like '1st', '2nd', '3rd', '4th'
def clean_date(date_str):
    if pd.isna(date_str):
        return date_str
    return re.sub(r'(\d+)(st|nd|rd|th)', r'\1', date_str)

# Apply cleaning
df['review_date'] = df['review_date'].astype(str).apply(clean_date)
df['date_of_travel'] = df['date_of_travel'].astype(str).apply(clean_date)

# Convert to datetime with a specified format
df['review_date'] = pd.to_datetime(df['review_date'], format="%d %B %Y", errors='coerce')
df['date_of_travel'] = pd.to_datetime(df['date_of_travel'], format="%B %Y", errors='coerce')

# Check the output
print(df[['review_date', 'date_of_travel']].head())
```
